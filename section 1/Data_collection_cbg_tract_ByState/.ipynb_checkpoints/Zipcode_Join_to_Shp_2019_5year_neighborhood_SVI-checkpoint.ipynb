As a result, you should have 5 json files: Zip_1_2019.json, Zip_2_2019.json, Zip_3_2019.json, Zip_4_2019.json, Zip_5_2019.json, Zip_6_2019.json

Now I am going to join the five json files. The data contains all zipcode in the U.S. But I need the data only for Illiois in this case, so I am going to join the data to the Illinois at the end.

```python
#set column header
Zip_1 = pd.read_json('Zip_1_2019.json')
new_header = Zip_1.iloc[0] 
Zip_1 = Zip_1[1:] 
Zip_1.columns = new_header
Zip_1
```

```python
#set column header
Zip_2 = pd.read_json('Zip_2_2019.json')
new_header = Zip_2.iloc[0] 
Zip_2 = Zip_2[1:] 
Zip_2.columns = new_header
Zip_2
```

```python
#set column header
Zip_3 = pd.read_json('Zip_3_2019.json')
new_header = Zip_3.iloc[0] 
Zip_3 = Zip_3[1:] 
Zip_3.columns = new_header
Zip_3
```

```python
#set column header
Zip_4 = pd.read_json('Zip_4_2019.json')
new_header = Zip_4.iloc[0] 
Zip_4 = Zip_4[1:] 
Zip_4.columns = new_header
Zip_4
```

```python
#set column header
Zip_5 = pd.read_json('Zip_5_2019.json')
new_header = Zip_5.iloc[0] 
Zip_5 = Zip_5[1:] 
Zip_5.columns = new_header
Zip_5
```

```python
#set column header
Zip_6 = pd.read_json('Zip_6_2019.json')
new_header = Zip_6.iloc[0] 
Zip_6 = Zip_6[1:] 
Zip_6.columns = new_header
Zip_6
```

```python
#join
tmp = Zip_1.set_index('zip code tabulation area').join(Zip_2.set_index('zip code tabulation area'), lsuffix = '_left', rsuffix = '_right')
tmp=tmp.reset_index()
tmp
```

```python
#join
tmp = tmp.set_index('zip code tabulation area').join(Zip_3.set_index('zip code tabulation area'))
tmp=tmp.reset_index()
tmp
```

```python
tmp = tmp.set_index('zip code tabulation area').join(Zip_4.set_index('zip code tabulation area'), lsuffix = '_left', rsuffix = '_right')
tmp=tmp.reset_index()
tmp
```

```python
#join
tmp = tmp.set_index('zip code tabulation area').join(Zip_5.set_index('zip code tabulation area'), lsuffix = '_left', rsuffix = '_right')
tmp=tmp.reset_index()
tmp
```

```python
tmp = tmp.set_index('zip code tabulation area').join(Zip_6.set_index('zip code tabulation area'), lsuffix = '_left', rsuffix = '_right')
tmp=tmp.reset_index()
tmp
```

```python
ACS_zipcode=tmp
#ACS_zipcode.drop(columns = ['index'], axis=1, inplace=True) 
ACS_zipcode.rename(columns = {'zip code tabulation area':'zipcode'}, inplace = True)
ACS_zipcode
```

```python
#set index
ACS_zipcode.set_index('zipcode',inplace=True)
ACS_zipcode
```

```python
ACS_zipcode.to_csv('ACS_2019_5year__zipcode_US.csv') # for all US
```
