### The following code will combine all the downloaded JSON files and save the result as a CSV file.

````python
import pandas as pd

# List of JSON files to load
files = [
    'cbg_1.json',
    'cbg_2.json',
    'cbg_3.json',
    'cbg_4.json',
    'cbg_5.json'
]

# Load all JSON files into DataFrames
dataframes = [pd.read_json(file) for file in files]

# Adjust column headers and drop the first row
adjusted_dataframes = []
for df in dataframes:
    df.columns = df.iloc[0]  # Set the first row as the header
    adjusted_df = df[1:]     # Drop the first row
    adjusted_dataframes.append(adjusted_df)

# Define the essential columns for merging
essential_columns = ["state", "county", "tract", "block group"]

# Prepare data with only essential columns and optimize data types
optimized_dataframes = []
for df in adjusted_dataframes:
    # Remove duplicate columns and keep all unique ones
    df = df.loc[:, ~df.columns.duplicated()]
    # Optimize data types
    for col in df.select_dtypes(include=["object"]).columns:
        df[col] = df[col].astype("category")
    optimized_dataframes.append(df)

# Merge the DataFrames iteratively
merged_data = optimized_dataframes[0]
for df in optimized_dataframes[1:]:
    merged_data = merged_data.merge(
        df,
        on=essential_columns,
        how="inner",
        suffixes=('', '_drop')
    )
    # Drop duplicate columns (those with '_drop' suffix)
    merged_data = merged_data[[col for col in merged_data.columns if not col.endswith('_drop')]]

# Save the merged DataFrame to a CSV file
output_file_path = '../../section 2/data/ACS_2023_5year__cbg_CA.csv'
merged_data.to_csv(output_file_path, index=False)
````


## Explanation of Command Arguments

### `-i` (Input File)
- Provide the combined ACS data file downloaded in the previous step. This file is the output generated during the earlier process.

### `-c` (Codebook File)
- Specify the Excel file containing the ACS variable codes and descriptions. Key columns in the codebook:
  - **`CODE`**: Contains the ACS variable codes. Multiple codes separated by commas in this column are summed and divided by the value in the **`denominator`** column.
  - **`use`**: Indicates whether the variable should be included in the output. Use `1` to include or leave it empty to exclude.
  - **`denominator`**: Optional: enter the variable code that you want to use to normalize the variable in ColA. If colC is empty, it will use the count (the value in  **`CODE`** column) without the normalization.
  - **`Short_name`**: These names will be used as column headers in the summarized output file (`XXXX_summarized.csv`).
  - **`Name_normalized`**: These names will be used as column headers in the normalized output file (`XXXX_normalized.csv`).

#### Output Breakdown
- **`XXXX_summarized.csv`**: Contains raw values or the sum of raw values when multiple codes are listed in the `CODE` column.
- **`XXXX_normalized.csv`**: Contains normalized values calculated as `CODE` / `denominator`.

### `-o` (Output Directory)
- Specify the directory where the output files will be saved.
