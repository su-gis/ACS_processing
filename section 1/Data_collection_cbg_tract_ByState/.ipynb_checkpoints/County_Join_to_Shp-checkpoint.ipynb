As a result, you should have 5 json files: County_1.json, County_2.json, County_3.json, County_4.json, County_5.json, County_6.json

```python
#set column header
County_1 = pd.read_json('County_1.json')
new_header = County_1.iloc[0] 
County_1 = County_1[1:] 
County_1.columns = new_header
County_1
```

```python
#set column header
County_2 = pd.read_json('County_2.json')
new_header = County_2.iloc[0] 
County_2 = County_2[1:] 
County_2.columns = new_header
County_2
```

```python
#set column header
County_3 = pd.read_json('County_3.json')
new_header = County_3.iloc[0] 
County_3 = County_3[1:] 
County_3.columns = new_header
County_3
```

```python
#set column header
County_4 = pd.read_json('County_4.json')
new_header = County_4.iloc[0] 
County_4 = County_4[1:] 
County_4.columns = new_header
County_4
```

```python
#Tract_5 column header
County_5 = pd.read_json('County_5.json')
new_header = County_5.iloc[0] 
County_5 = County_5[1:] 
County_5.columns = new_header
County_5
```

```python
#set column header
County_6 = pd.read_json('County_6.json')
new_header = County_6.iloc[0] 
County_6 = County_6[1:] 
County_6.columns = new_header
County_6
```

```python
#tmp
tmp = County_1.merge(County_2, how='inner', left_on=["state", "county"], right_on=["state", "county"])
tmp
```

```python
#tmp
tmp = tmp.merge(County_3, how='inner', left_on=["state", "county"], right_on=["state", "county"])
tmp
```

```python
#tmp
tmp = tmp.merge(County_4, how='inner', left_on=["state", "county"], right_on=["state", "county"])
tmp
```

```python
#tmp
tmp = tmp.merge(County_5, how='inner', left_on=["state", "county"], right_on=["state", "county"])
tmp
```

```python
#tmp
tmp = tmp.merge(County_6, how='inner', left_on=["state", "county"], right_on=["state", "county"])
tmp
```

```python
tmp.to_csv('ACS_2018_5year__County_IL.csv')
```
