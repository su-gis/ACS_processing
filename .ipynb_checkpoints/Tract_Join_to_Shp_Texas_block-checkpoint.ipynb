As a result, you should have 5 json files: Tract_1.json, Tract_2.json, Tract_3.json, Tract_4.json, Tract_5.json, Trac_6.json

Now I am going to join the five json files. 

```python
#set column header
Tract_1 = pd.read_json('TX_block1.json')
new_header = Tract_1.iloc[0] 
Tract_1 = Tract_1[1:] 
Tract_1.columns = new_header
Tract_1
```

```python
#set column header
Tract_2 = pd.read_json('TX_block2.json')
new_header = Tract_2.iloc[0] 
Tract_2 = Tract_2[1:] 
Tract_2.columns = new_header
Tract_2
```

```python
#set column header
Tract_3 = pd.read_json('TX_block3.json')
new_header = Tract_3.iloc[0] 
Tract_3 = Tract_3[1:] 
Tract_3.columns = new_header
Tract_3
```

```python
#set column header
Tract_4 = pd.read_json('TX_block4.json')
new_header = Tract_4.iloc[0] 
Tract_4 = Tract_4[1:] 
Tract_4.columns = new_header
Tract_4
```

```python
#Tract_5 column header
Tract_5 = pd.read_json('TX_block5.json')
new_header = Tract_5.iloc[0] 
Tract_5 = Tract_5[1:] 
Tract_5.columns = new_header
Tract_5
```

```python
#tmp
tmp = Tract_1.merge(Tract_2, how='inner', left_on=["state", "county", "tract", "block group"], right_on=["state", "county", "tract", "block group"])
tmp
```

```python
#join
tmp = tmp.merge(Tract_3, how='inner', left_on=["state", "county", "tract", "block group"], right_on=["state", "county", "tract", "block group"])
tmp
```

```python
tmp = tmp.merge(Tract_4, how='inner', left_on=["state", "county", "tract", "block group"], right_on=["state", "county", "tract", "block group"])
tmp
```

```python
tmp = tmp.merge(Tract_5, how='inner', left_on=["state", "county", "tract", "block group"], right_on=["state", "county", "tract", "block group"])
tmp
```

```python
tmp.to_csv('ACS_2019_5year__block_Texas.csv')
```
